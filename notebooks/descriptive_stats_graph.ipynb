```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
pd.set_option("display.max_columns", None)
```

```python
data = pd.read_csv('gs://websitehumantrafficking/final_df.csv')
data
```

```python
data['ageBroad'].notna()
```

```python
data['ageBroad'].replace('Unknown', np.nan, inplace = True)
```

```python
data['ageBroad'].replace('nan', np.nan, inplace = True)
```

```python
data['Victims'] = 1
fig = px.sunburst(data[data.ageBroad.notna()], path=['gender', 'ageBroad'], values='Victims', color='gender',
                  title='Gender and Age of Human Trafficking Victims')
fig.update_layout(width=600, height=600)
fig.show()
```
