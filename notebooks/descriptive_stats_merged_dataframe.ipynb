```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import numpy as np
```

```python
HT_df = pd.read_csv('gs://websitehumantrafficking/final_df.csv')
HT_df
```

```python
cm = sns.light_palette("red", as_cmap=True)
table = pd.pivot_table(HT_df, values='Datasource', index='arrival_1',
                    columns='year', aggfunc='count', fill_value=0)
table.style.background_gradient(cmap=cm)
```

```python
HT_df['gender'] =HT_df['gender'].astype(str)
indexGender = HT_df[ HT_df['gender'] == 'nan'].index
HT_df.drop(indexGender , inplace=True)
HT_df['gender'].unique()
```

```python
HT_df['ageBroad'] =HT_df['ageBroad'].astype(str)
indexAge = HT_df[ HT_df['ageBroad'] == 'nan'].index
HT_df.drop(indexAge , inplace=True)
HT_df['ageBroad'].unique()
```

```python
HT_df['Victims'] = 1
fig = px.sunburst(HT_df[HT_df.ageBroad.notna()], path=['gender', 'ageBroad'], values='Victims', color='gender',
                  title='Gender and Age of Human Trafficking Victims')
fig.update_layout(width=600, height=600)
fig.show()
```
