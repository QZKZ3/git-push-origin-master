```python
df_localization2['arrival_1'].astype(str)
```

```python
def strip_function(item):
        try:
            return item.strip()
        except AttributeError:
            pass
   
df_localization2['arrival_1'] = df_localization2['arrival_1'].apply(strip_function)
```

```python
df_localization2['arrival_2'] = df_localization2['arrival_2'].apply(strip_function)
```

```python
df_localization2['arrival_3'] = df_localization2['arrival_3'].apply(strip_function)
```

```python
df_localization2['arrival_1'].iloc[4]
```

```python
def localization_cleaning(item):
    if (item.country==item['departure']) | (item.country==item['arrival_1']) | (item.country==item['arrival_2']) | (item.country==item['arrival_3']):
        return item
    else:
        return 0
```

```python
df_localization2['localization_cleaning'] = df_localization2.apply(localization_cleaning,axis=1)
df_localization2
```

```python
df_localization2.drop(df_localization2 [df_localization2.localization_cleaning == 0], inplace=True)
```
