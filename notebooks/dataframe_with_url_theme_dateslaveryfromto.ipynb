```python
def convert_to_datatime(n):
    if len(n)==9:
        return [n[0:4],n[5:9]]
    else:
        return [np.nan, np.nan]
```

```python
df_final_with_url_and_column_themes['date_slavery'] = df_final_with_url_and_column_themes['date_slavery'].apply(convert_to_datatime)
```
